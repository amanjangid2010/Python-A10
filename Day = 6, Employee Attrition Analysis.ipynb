# Employee Attrition Analysis

#Employee attrition analysis is a type of behavioural analysis where we study the behaviour and
#characteristics of the employees who left the organization and compare their characteristics with
#The current employees to find the employees who may leave the organization soon.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Attrition - Attrition.csv")
df.head()
```

```python
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
print(df.isnull().sum())
```

```python
# Check percentage of Attrition by department 
```

```python
# Filter the data to show only "Yes" values in the "Attrition" Column
```

```python
attr_df = df[df["Attrition"]=="Yes"]
attr_df
```

```python
# Calculate the attrition by departemet
```

```python
attrition_by_dpt = attr_df.groupby(["Department"]).size().reset_index(name = "count")
attrition_by_dpt
```

```python
# Create a donut chart

fig = go.Figure(data=[go.Pie(
    labels = attrition_by_dpt["Department"],
    values = attrition_by_dpt["count"],
    hole = 0.4,
    marker = dict(colors=["red", "green"]),
    textposition = "inside"
)])

# Update the layout
fig.update_layout(title="Attrition by Department")

# Show the Chart
fig.show()
```

```python
attrition_by = attr_df.groupby(["EducationField"]).size().reset_index(name="count")
attrition_by
```

```python
# Create a donut chart

fig = go.Figure(data=[go.Pie(
    labels = attrition_by["EducationField"],
    values = attrition_by['count'],
    hole = 0.4,
    marker = dict(colors=["red", "green"]),
    textposition = "inside"
)])

# Update the layout
fig.update_layout(title="Attrition by EducationField")

# Show the Chart
fig.show()
```

# We can see that the employees with Life Sciences as an education field have a hight attrition rate. Now let's have a look at the percentage of attrition by number of years at the company

```python
attrition_by = attr_df.groupby(["YearsAtCompany"]).size().reset_index(name="Count")
attrition_by
```

```python
# We can see that most of the employees leave the organization after completing a year.
```

```python
attrition_by = attr_df.groupby(["YearsSinceLastPromotion"]).size().reset_index(name="Count")
attrition_by
```

# Now let's have a look at the percentage of attrition by gender:

```python
attrition_by = attr_df.groupby(["Gender"]).size().reset_index(name="count")
fig = go.Figure(data=[go.Pie(
    labels = attrition_by["Gender"],
    values = attrition_by['count'],
    hole = 0.4,
    marker = dict(colors=["red", "green"]),
    textposition = "inside"
)])

# Update the layout
fig.update_layout(title="Attrition by Gender")

# Show the Chart
fig.show()
```
